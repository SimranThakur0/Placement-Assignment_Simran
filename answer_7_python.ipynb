```python
import requests
import json
import matplotlib.pyplot as plt

def download_data(url):
    # Send a GET request to the URL to download the data
    response = requests.get(url)
    # Check if the request was successful
    if response.status_code == 200:
        # Return the downloaded data
        return response.text
    else:
        # If the request failed, raise an exception
        raise Exception(f"Failed to download data. Error code: {response.status_code}")

def process_data(data):
    # Parse the JSON data
    meteorites = json.loads(data)
    return meteorites

def filter_meteorites_by_year(meteorites, year):
    # Filter meteorites based on the specified year
    filtered_meteorites = [meteorite for meteorite in meteorites if meteorite.get('year', '') < year]
    return filtered_meteorites

def filter_meteorites_by_mass(meteorites, mass_threshold):
    # Filter meteorites based on mass (assuming mass is in kg)
    filtered_meteorites = [meteorite for meteorite in meteorites if meteorite.get('mass (g)', 0) / 1000 > mass_threshold]
    return filtered_meteorites

def extract_coordinates(meteorites):
    # Extract coordinates from meteorites data
    coordinates = [(meteorite.get('reclat', ''), meteorite.get('reclong', '')) for meteorite in meteorites]
    return coordinates

def plot_coordinates(coordinates):
    # Unpack latitude and longitude values from the coordinates list
    latitudes, longitudes = zip(*coordinates)

    # Create a scatter plot of the coordinates
    plt.scatter(longitudes, latitudes, s=5, alpha=0.7)
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title('Earth Meteorites Coordinates')
    plt.show()

# URL to download the data from
url = "https://data.nasa.gov/resource/y77d-th95.json"

# Download the data from the URL
data = download_data(url)

# Process the data
meteorites = process_data(data)

# Get all the Earth meteorites that fell before the year 2000
meteorites_before_2000 = filter_meteorites_by_year(meteorites, 2000)

# Get all the Earth meteorites coordinates that fell before the year 1970
meteorites_coordinates_before_1970 = filter_meteorites_by_year(meteorites, 1970)
coordinates_before_1970 = extract_coordinates(meteorites_coordinates_before_1970)

# Assuming mass is in kg, get all Earth meteorites with mass > 10000 kg
meteorites_mass_gt_10000kg = filter_meteorites_by_mass(meteorites, 10000)
mass_gt_10000kg_coordinates = extract_coordinates(meteorites_mass_gt_10000kg)

# Plot the coordinates of all Earth meteorites
plot_coordinates(extract_coordinates(meteorites))

# Plot the coordinates of Earth meteorites before 1970
plot_coordinates(coordinates_before_1970)

# Plot the coordinates of Earth meteorites with mass > 10000 kg
plot_coordinates(mass_gt_10000kg_coordinates)
```
